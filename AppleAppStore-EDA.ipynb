Fill Price and Size_Bytes by mean

```python
df['Price'] = df['Price'].fillna(value=df['Price'].mean())
df['Size_Bytes'] = df['Size_Bytes'].fillna(value=df['Size_Bytes'].mean())
```

Drop 'Developer_Url','Released','App_Name' Null Rows

```python
df.dropna(subset=['Developer_Url','Released','App_Name'],inplace=True)
```

```python
round(df.isnull().sum()*100/len(df)).sort_values(ascending=False)
```

Checking Content_Rating unique values

```python
df['Content_Rating'].unique()
```

```python
df['Content_Rating'] = df['Content_Rating'].replace('4+',"Children")
df['Content_Rating'] = df['Content_Rating'].replace('9+',"School Childs")
df['Content_Rating'] = df['Content_Rating'].replace('12+',"Teens")
df['Content_Rating'] = df['Content_Rating'].replace('17+',"Adults")
df['Content_Rating'] = df['Content_Rating'].replace('Not yet rated',"Everyone")
```

```python
df['Content_Rating'].unique()
```

Rename Content_Rating with Age_Group

```python
df.rename(columns={'Content_Rating':'Age_Group'},inplace=True)
```

```python
df.head(5)
```

```python
df['Reviews'].max()
```

Total Reviews are 22685334

```python
bins = [-1,0,10000,500000,1000000,22685334]
labels = ['No Reviews','Less than 10K','Less than 50K','Less than 1M','1M+']
df['Review_Category'] = pd.cut(df['Reviews'],bins=bins,labels=labels)
```

### **Data Insights**

Question 1. What are the top 10 Categories that are installed from the Apple Store ?



```python
top10 = df['Primary_Genre'].value_counts().reset_index().sort_values(by='count',ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(16,6))
sns.barplot(top10,x=top10['Primary_Genre'],y=top10['count'])
```

> Games have highest number of Installation

Question 2. What are the highest top 10 rated primary_genre based on Average_User_Rating

```python
df.groupby('Primary_Genre')['Average_User_Rating'].mean().sort_values(ascending=False).head(10)
```

> Weather have highest rating

Question 3. Which Primary_Genre has the highest count of Paid and Free apps ?

```python
Category_Type = df.groupby('Primary_Genre')['Type'].value_counts().sort_values(ascending=False).reset_index(name='Count')
```

```python
Category_Free = Category_Type[Category_Type['Type']=='Free']
Category_Paid = Category_Type[Category_Type['Type']=='Paid']
```

```python
Category_Free.head(10)
```

> Games have highest frequency of Free Apps

```python
Category_Paid.head(10)
```

> Education have highest frequency of Paid Apps

Question 4. What are the Top 5 Paid Apps based with highest ratings?

```python
top5Paid = df[df['Type']== 'Paid'].sort_values('Average_User_Rating',ascending=False)
top5Paid.head(5)[['App_Name','Average_User_Rating']]
```

> Super Nano Trucks Free App have highest Rating

Question 5. What are the Top 5 Free Apps based with highest ratings?

```python
top5Free = df[df['Type']== 'Free'].sort_values('Average_User_Rating',ascending=False)
top5Free.head(5)[['App_Name','Average_User_Rating']]
```

> Rise of Zombie - City Defense Paid App have highest Rating

Question 6. Apps with highest Age_Group

```python
df['Age_Group'].value_counts()
```

> Children Group have highest no of Apps

Question 7. Years in which max apps were released

```python
# Convert Released Column to Date and time format
df['Released'] = pd.to_datetime(df['Released'])
df['Released'].head(5)
```

```python
# Create a Column Year_Released, then save only year in this column
df['Year_Released'] = df['Released'].dt.strftime('%Y')
df['Year_Released'].head(5)
```

```python
year_release = df['Year_Released'].value_counts().reset_index(name='count').sort_values(by='count',ascending=False).head(10)
year_release
```

```python
sns.barplot(year_release,x='Year_Released',y='count')
```

> 2020 is the year when highest number of Apps Released

Question 8. Size in MBs vs. Price of App

```python
Size_Apps_Price = df.groupby('Size_Bytes')['Price'].value_counts().reset_index(name='count').sort_values(by='Size_Bytes',ascending=False).head(5)
Size_Apps_Price
```

```python
plt.figure(figsize=(26,6))
sns.barplot(Size_Apps_Price,x='Size_Bytes',y='Price')
```

Question 9. Top 10 app producing developer

```python
top10Dev = df['Developer'].value_counts().reset_index(name='count').sort_values(by='count',ascending=False).head(10)
top10Dev
```

```python
plt.figure(figsize=(15,5))
data = top10Dev.groupby('Developer')['count'].max().sort_values(ascending = True)
data = data.head(10)
labels = data.keys()
plt.pie(data, labels= labels,autopct='%.0f%%') 
plt.title(" Top 10 App Producing developer", fontsize=14)
```

Question 10. YoY (Year on Year) comparison of apps per Age_Group

```python
yoy = df.groupby('Age_Group')['Year_Released'].value_counts().reset_index(name='count')
plt.figure(figsize=(26,6))
sns.barplot(x=yoy['Year_Released'],y=yoy['count'],hue=yoy['Age_Group'])
```
